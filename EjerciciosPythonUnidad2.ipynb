2.1 Realice diagramas Pv de isotermas para gases que obedecen las siguientes ecuaciones de estado
(realice los diagramas de las isotermas para distintos valores de a y b y compárelas):
(a) Ecuación del gas ideal
(b) Ecuación de van der Waals
(c) Redlich-Kwong
(d) Ecuación de Dieterici
(e) Ecuación de Berthelot

```python
R = 8.314472 #Gas constant (J/K*mol)
```

```python
#Define functions to plot
def P1(T):
    v = np.arange(0.1,5.,0.01)
    plt.figure(figsize=(10,5))
    plt.plot(v, R*T/v)
    plt.title('IDEAL GAS EQUATION ISOTHERM', size=15)
    plt.ylabel(r'$P \left ( \frac{N}{m^{2}} \right )$', size=15)
    plt.xlabel(r'$V \left ( m^{3} \right )$', size=15)
    plt.grid()

def P2(T,a,b):
    v=np.arange(b+0.1,5.,0.01)
    p2=(b*8*T/(v-b))-(a/v**2)
    plt.figure(figsize=(10,5))
    plt.plot([b]*len(np.arange(np.min(p2),np.max(p2)+0.01,0.01)),np.arange(np.min(p2),np.max(p2)+0.01,0.01),"r--")
    plt.plot(v, p2)
    plt.title('VAN DER WAALS EQUATION ISOTHERM', size=15)
    plt.ylabel(r'$P \left ( \frac{N}{m^{2}} \right )$', size=15)
    plt.xlabel(r'$V \left ( m^{3} \right )$', size=15)
    plt.ylim(np.min(p2),np.max(p2))
    plt.grid()
    
def P3(T,a,b):
    v=np.arange(b+0.1,5.,0.01)
    p3=(R*T/(v-b))-(a/(v*(v+b)*T**(1/2)))
    plt.figure(figsize=(10,5))
    plt.plot([b]*len(np.arange(np.min(p3),np.max(p3)+0.01,0.01)),np.arange(np.min(p3),np.max(p3)+0.01,0.01),"r--")
    plt.plot(v, p3)
    plt.title('REDLICH-KWONG EQUATION ISOTHERM', size=15)
    plt.ylabel(r'$P \left ( \frac{N}{m^{2}} \right )$', size=15)
    plt.xlabel(r'$V \left ( m^{3} \right )$', size=15)
    plt.ylim(np.min(p3),np.max(p3))
    plt.grid()
    

def P4(T,a,b):
    v=np.arange(b+0.1,5.,0.01)
    p4=(R*T/(v-b))*np.exp(-a/R*T*v)
    plt.figure(figsize=(10,5))
    plt.plot([b]*len(np.arange(np.min(p4),np.max(p4)+0.01,0.01)),np.arange(np.min(p4),np.max(p4)+0.01,0.01),"r--")
    plt.plot(v, p4)
    plt.title('DIETERICI EQUATION ISOTHERM', size=15)
    plt.ylabel(r'$P \left ( \frac{N}{m^{2}} \right )$', size=15)
    plt.xlabel(r'$V \left ( m^{3} \right )$', size=15)
    plt.ylim(np.min(p4),np.max(p4))
    plt.grid()

def P5(T,a,b):
    v=np.arange(b+0.1,5.,0.01)
    p5=(b*8*T/(v-b))-(a/(T*v**2))
    plt.figure(figsize=(10,5))
    plt.plot([b]*len(np.arange(np.min(p5),np.max(p5)+0.01,0.01)),np.arange(np.min(p5),np.max(p5)+0.01,0.01),"r--")
    plt.plot(v, p5)
    plt.title('BERTHELOT EQUATION ISOTHERM', size=15)
    plt.ylabel(r'$P \left ( \frac{N}{m^{2}} \right )$', size=15)
    plt.xlabel(r'$V \left ( m^{3} \right )$', size=15)
    plt.ylim(np.min(p5),np.max(p5))
    plt.grid()
```

IDEAL GAS EQUATION

$P = \frac{RT}{V}$


```python
fig, ax = plt.subplots(figsize=(10,5))

#Ideal gas equation isothermtherms
for T in np.arange(0.1,2,0.2):
    v = np.arange(0.01,6.,0.01)
    ax.plot(v, R*T/v)
    

ax.set_title('IDEAL GAS EQUATION ISOTHERMS', size=15)
ax.set(xlabel=r'$V \left ( m^{3} \right )$', ylabel=r'$P \left ( \frac{N}{m^{2}} \right )$')
ax.set_ylim(-1,30)
ax.set_xlim(-0.3,5)
ax.grid()
plt.show()
```

```python
interact(P1,T=1)
plt.show()
```

VAN DER WAALS EQUATION

$P = \frac{RT}{V-b} - \frac{a}{V^{2}}$

```python
#Van Der Waals Equation isotherms for a=3 and b=1/3
a=3; b=1/3
v=np.arange(b+0.01,5.5,0.01)
    
fig, ax1 = plt.subplots(figsize=(10,5))

#Ideal gas equation isothermtherms
for T in np.arange(0.1,2,0.2):
    ax1.plot(v, (b*8*T/(v-b))-(a/v**2))

ax1.plot([b]*len(np.arange(-16,21,1)),np.arange(-16,21,1),"r--")
ax1.set_title('VAN DER WAALS EQUATION ISOTHERMS', size=15)
ax1.set(xlabel=r'$V \left ( m^{3} \right )$', ylabel=r'$P \left ( \frac{N}{m^{2}} \right )$')
ax1.grid()
ax1.text(3,16,r'$P = \frac{RT}{V-b} - \frac{a}{V^{2}}$   a = 3; b = 1/3',fontsize=15)
ax1.set_ylim(-16,20)
ax1.set_xlim(0,5)
plt.show()
```

```python
interact(P2,T=(0.1,2.0,0.1),a=3,b=1/3)
plt.show()
```

REDLICH-KWONG EQUATION

$P = \frac{RT}{V-b} - \frac{aT^{-0.5}}{V(V-b)}$


```python
#Redlich-Kwong Equation isotherms for a=0.42748*R**2; b=0.08664*R
a=0.42748*R**2; b=0.08664*R
v=np.arange(b+0.01,5.,0.01)  
fig, ax2 = plt.subplots(figsize=(10,5))

#Ideal gas equation isothermtherms
for T in np.arange(0.3,2.1,0.2):
    ax2.plot(v, (R*T/(v-b))-(a/(v*(v+b)*T**(1/2))))


ax2.plot([b]*len(np.arange(-24,21,1)),np.arange(-24,21,1),"r--")
ax2.set_title('REDLICH-KWONG EQUATION ISOTHERM', size=15)
ax2.set(xlabel=r'$V \left ( m^{3} \right )$', ylabel=r'$P \left ( \frac{N}{m^{2}} \right )$')
ax2.text(1.9,16,r'$P = \frac{RT}{V-b} - \frac{aT^{-0.5}}{V(V-b)}$   $a=0.42748R^{2} $; b=0.08664*R ',fontsize=15)
ax2.grid()

ax2.set_xlim(0.5,5)
ax2.set_ylim(-24,20)
plt.show()
```

```python
interact(P3,T=(0.1,2.0,0.1),a=0.42748*R**2, b=0.08664*R)
plt.show()
```

DIETERICI EQUATION

$P = \frac{RT}{V-b}e^{-a/RTV}$

```python
#Dieterici Equation isotherms for a=2*R  b=1/2
a=2*R

b=1/2
v=np.arange(b+0.01,5.5,0.01)
    
fig, ax3 = plt.subplots(figsize=(10,5))

for T in np.arange(0.1,2.1,0.3):
    ax3.plot(v, (R*T/(v-b))*np.exp(-a/R*T*v))


ax3.plot([b]*len(np.arange(-21,21,1)),np.arange(-21,21,1),"r--")
ax3.set_title('DIETERICI EQUATION ISOTHERMS FOR 0<T<2', size=15)
ax3.set(xlabel=r'$V \left ( m^{3} \right )$', ylabel=r'$P \left ( \frac{N}{m^{2}} \right )$')
ax3.grid()
ax3.set_ylim(0,10)
ax3.set_xlim(0,5)
plt.show()
```

```python
interact(P4,T=(0.1,2.0,0.1),a=2*R, b=1/2)
plt.show()
```

```python
#Bethelot Equation isotherms for a=3 and b=1/3
a=3; b=1/3
v=np.arange(b+0.01,5.5,0.01)
fig, ax4 = plt.subplots(figsize=(10,5))

#Ideal gas equation isothermtherms
for T in np.arange(0.1,2,0.2):
    ax4.plot(v, (b*8*T/(v-b))-(a/(T*v**2)))

ax4.plot([b]*len(np.arange(-25,21,1)),np.arange(-25,21,1),"r--")
ax4.set_title('BERTHELOT ISOTHERM FOR 0<T<2', size=15)
ax4.set(xlabel=r'$V \left ( m^{3} \right )$', ylabel=r'$P \left ( \frac{N}{m^{2}} \right )$')
ax4.grid()
ax4.text(3,16,r'$P = \frac{RT}{V-b} - \frac{a}{TV^{2}}$   a = 3; b = 1/3',fontsize=15)
ax4.set_ylim(-25,20)
ax4.set_xlim(0,5)
plt.show()
```

2.2 Realice un estudio (grafique e interprete) de la función de distribución de velocidades de
Maxwell-Boltzmann, f(v), para distintos valores (cuatro como mínimo) de la masa m de las
partículas y de la temperatura, T. 
Donde v es la velocidad de la partícula y k es la constante de Boltzmann (1.381x10 -23 J/K.).

```python
k=1.381e-23
m=[1.66e-27,3.31982506e-26,6.635914366e-26,1.293898981e-25, 2.190321493e-25]
names=[r'$^{4}He$',r'$^{20}Ne$',r'$^{40}Ar$',r'$^{78}Kr$',r'$^{132}Xe$']

def F(T):
    v = np.arange(0,6000,0.10)
    plt.figure(figsize=(10,5))
    for i in np.arange(0,5,1):
        f=4*np.pi*v**2*((m[i]/(2*np.pi*k*T))**(3/2))*np.exp(-m[i]*(v)**2/(2*k*T))
        plt.plot(v, f, label=names[i])
        plt.xlim
    plt.title('Maxwell-Boltzman Molecular Speed Distribution',fontsize=20)
    plt.ylabel('Probability Density',fontsize=20)
    plt.xlabel('Speed(m/s)',fontsize=20)
    plt.grid()
    plt.legend(fontsize=15)


        
interact(F,T=(50,1000,50))


plt.show()

```
